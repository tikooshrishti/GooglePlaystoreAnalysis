```python
duplicate_apps = pd.read_csv("apps.csv")
```

```python
apps = duplicate_apps.drop_duplicates()
```

```python
print('Total no. of apps in the dataset = ','apps')
print(apps.head())
```

```python
n = 7
apps.sample(n)
```

```python
#Data Cleaning
chars_to_remove = ['+',',','$']
cols_to_clean = ['Installs','Price']
```

```python
for col in cols_to_clean:
    for char in chars_to_remove:
        apps[col] = apps[col].astype(str).str.replace(char,'')
    apps[col] = pd.to_numeric(apps[col])
```

```python
num_categories = len(apps['Category'].unique())
print('Number of categories = ', num_categories)
```

```python
num_apps_in_category = apps['Category'].value_counts().sort_values(ascending = False)

data = [go.Bar(
        x = num_apps_in_category.index,
        y = num_apps_in_category.values,
)]

plotly.offline.iplot(data)
```

```python
#App Ratings
avg_app_rating = apps['Rating'].mean()
print('Average app rating = ', avg_app_rating)
```

```python
#Histogram
data = [go.Histogram(
        x = apps['Rating']
)]
layout = {'shapes': [{
              'type' :'line',
              'x0': avg_app_rating,
              'y0': 0,
              'x1': avg_app_rating,
              'y1': 1000,
              'line': { 'dash': 'dashdot'}
          }]
          }

plotly.offline.iplot({'data': data, 'layout': layout})
```

```python
#App category Vs App Price
import matplotlib.pyplot as plt
fig,ax = plt.subplots()
fig.set_size_inches(13,7)
```

```python
popular_app_cata = apps[apps.Category.isin(['GAME', 'FAMILY', 'PHOTOGRAPHY',
                                            'MEDICAL', 'TOOLS', 'FINANCE',
                                            'LIFESTYLE','BUSINESS'])]
```

```python
ax = sns.stripplot(x = popular_app_cata['Price'], y = popular_app_cata['Category'], jitter=True, linewidth=1)
ax.set_title('App pricing trend across categories')
```

```python
#Prices Greater than 200
apps_above_200 = popular_app_cata[['Category', 'App', 'Price']][popular_app_cata['Price'] > 200]
apps_above_200
```

```python
#prices below 100
apps_under_100 = popular_app_cata[popular_app_cata['Price'] < 100]

fig, ax = plt.subplots()
fig.set_size_inches(13, 7)
```

```python
ax = sns.stripplot(x='Price', y='Category', data=apps_under_100,
                   jitter=True, linewidth=1)
ax.set_title('App pricing trend across categories after filtering for junk apps')
```

```python
#paid vs free apps
trace0 = go.Box(
    y=apps[apps['Type'] == 'Paid']['Installs'],
    name = 'Paid'
)
```

```python
trace1 = go.Box(
    # Data for free apps
    y=apps[apps['Type'] == 'Free']['Installs'],
    name = 'Free'
)
```

```python
layout = go.Layout(
    title = "Number of downloads of paid apps vs. free apps",
    yaxis = dict(
        type = 'log',
        autorange = True
    )
)
```

```python
data = [trace0, trace1]
plotly.offline.iplot({'data': data, 'layout': layout})
```
